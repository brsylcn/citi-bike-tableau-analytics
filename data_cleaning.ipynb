```python
#Import Dependencies

import pandas as pd
import numpy as np
import datetime as dt
from datetime import date
```

```python
# Rename the csv files to make them ready to read

data01 = "data/JC-201901-citibike-tripdata.csv"
data02 = "data/JC-201902-citibike-tripdata.csv"
data03 = "data/JC-201903-citibike-tripdata.csv"
data04 = "data/JC-201904-citibike-tripdata.csv"
data05 = "data/JC-201905-citibike-tripdata.csv"
data06 = "data/JC-201906-citibike-tripdata.csv"
data07 = "data/JC-201907-citibike-tripdata.csv"
data08 = "data/JC-201908-citibike-tripdata.csv"
data09 = "data/JC-201909-citibike-tripdata.csv"
data10 = "data/JC-201910-citibike-tripdata.csv"
data11 = "data/JC-201911-citibike-tripdata.csv"
data12 = "data/JC-201912-citibike-tripdata.csv"
```

```python
# Read csv files

jan = pd.read_csv(data01)
feb = pd.read_csv(data02)
mar = pd.read_csv(data03)
apr = pd.read_csv(data04)
may = pd.read_csv(data05)
jun = pd.read_csv(data06)
jul = pd.read_csv(data07)
aug = pd.read_csv(data08)
sep = pd.read_csv(data09)
octo = pd.read_csv(data10)
nov = pd.read_csv(data11)
dec = pd.read_csv(data12)
```

```python
jan.head()
```

```python
octo.head()
```

```python
# Merge all files

year19 = jan.append(feb).append(mar).append(apr).append(may).append(jun).append(jul)\
         .append(aug).append(sep).append(octo).append(nov).append(dec)
```

```python
year19.head()
```

```python
year19 = year19.reset_index(drop=True)
year19
```

```python
# Replace the gender no with genders

year19['gender'].replace(to_replace=[0], value='n/a', inplace=True)
year19['gender'].replace(to_replace=[1], value='male', inplace=True)
year19['gender'].replace(to_replace=[2], value='female', inplace=True)
```

```python
year19.head()
```

```python
year19['age'] = 2020 - year19['birth year']
```

```python
# Create bins in which to place values based upon Age Demographics
bins = [0, 9.99, 19.99, 29.99, 39.99, 49.99, 59.99, 69.99, 79.99, 200]

# Create labels for these bins
group = ["<10", "10-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70-79", "80+"]
```

```python
# Slice the data and place it into bins
# Place the data series into a new column inside of the DataFrame
year19_age = year19["ageGroup"] = pd.cut(year19["age"], bins, labels = group)

year19_age
```

```python
# Save dataframe to csv file

year19.to_csv('year19_age.csv',mode = 'w', index=False)
```
